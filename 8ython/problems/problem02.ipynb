# 평균점수
> 다음 딕셔너리에서 평균 점수를 출력하라


````python
student = {'python':80, 'algorithm':78, 'django':95, 'flask':80}
def pyeonggyun(data):
    if type(data) == dict :
        pyengyun = sum(data.values())/len(data.keys())
        return pyengyun
    else :
        return 'Dictionary형 인자를 넣어주십시오.'
pyeonggyun(student)
````

# 혈액형
> 학생들의 혈액형(A,B,O,AB)에 대한 데이터가 리스트에 들어있다. 각 혈액형이 몇명인지 딕셔너리를 만들어 출력하라

````python
import collections
blood = ['A','A','B','O','A','B','A','AB','AB','O','A','O','AB','O']

def countman(data):
    
    return dict(collections.Counter(data))

countman(blood)
````

# UBD
> movies는 영화제목이 key로 누적관객수가 value인 딕셔너리이다. 
>
> 자전차왕 엄복동의 누적관객수는 172212명이고 172212명을 1UBD라고 할때 80UBD를 넘지 못하는 영화를 출력하라.

````python
movies = {
    "7번방의선물":12811206,
    "괴물":13019740,
    "국제시장":14257115,
    "극한직업":16261018,
    "도둑들":12983330,
    "명량":17613682,
    "베테랑":13414009,
    "신과함께-죄와벌":14410754,
    "아바타":13624328,
    "어벤져스:엔드게임":13901423,
}



def determinant_U(data):
    UBD = 172212
    for i in data.keys():
        if data.get(i) < 80 * UBD:
            print(f'영화 {i}은(는) 80UBD을 넘지 못한 영화입니다.')
    return

determinant_U(movies)
````
